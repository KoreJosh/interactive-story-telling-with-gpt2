```python
import torch
from transformers import GPT2Tokenizer, GPT2LMHeadModel

# Load pre-trained GPT-2 model and tokenizer
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')

# Function to generate text based on a prompt
def generate_text(prompt, max_new_tokens=50, temperature=0.7, no_repeat_ngram_size=2):
    input_ids = tokenizer.encode(prompt, return_tensors='pt')
    output = model.generate(
        input_ids,
        max_new_tokens=max_new_tokens,  # Use max_new_tokens instead of max_length
        temperature=temperature,
        no_repeat_ngram_size=no_repeat_ngram_size,
        pad_token_id=tokenizer.eos_token_id
    )
    return tokenizer.decode(output[0], skip_special_tokens=True)

# Function to simulate user input
def get_user_choice():
    return input("What do you want to do next? ")

# Main interactive storytelling function
def interactive_storytelling():
    print("Welcome to Interactive Storytelling!")
    print("You will guide the story by making choices at key moments.")

    # Initial story prompt
    story = "Once upon a time, in a small village, there lived a brave adventurer named Alex. One day, Alex decided to explore the mysterious forest nearby."
    print("\n" + story)

    # Loop for 5 user inputs
    for step in range(5):
        print(f"\n--- Step {step + 1} ---")
        user_choice = get_user_choice()

        # Update the story with the user's choice
        story += f" Alex decided to {user_choice}. "

        # Generate the next part of the story
        next_part = generate_text(story, max_new_tokens=50)  # Generate up to 50 new tokens
        print("\n" + next_part)

        # Update the story with the generated text
        story = next_part

    # End of the story
    print("\n--- The End ---")
    print("Thank you for playing! Here's your complete story:")
    print("\n" + story)

# Run the interactive storytelling
interactive_storytelling()
```

```python
import torch
from transformers import GPT2Tokenizer, GPT2LMHeadModel
import gradio as gr

# Load pre-trained GPT-2 model and tokenizer
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')

# Function to generate text based on a prompt
def generate_text(prompt, max_new_tokens=50, temperature=0.7, no_repeat_ngram_size=2):
    input_ids = tokenizer.encode(prompt, return_tensors='pt')
    output = model.generate(
        input_ids,
        max_new_tokens=max_new_tokens,  # Use max_new_tokens instead of max_length
        temperature=temperature,
        no_repeat_ngram_size=no_repeat_ngram_size,
        pad_token_id=tokenizer.eos_token_id
    )
    return tokenizer.decode(output[0], skip_special_tokens=True)

# Function to handle the interactive storytelling
def interactive_storytelling(user_choice, story_state=""):
    if not story_state:
        # Initial story prompt
        story = "Once upon a time, in a small village, there lived a brave adventurer named Alex. One day, Alex decided to explore the mysterious forest nearby."
    else:
        story = story_state

    # Update the story with the user's choice
    story += f" Alex decided to {user_choice}. "

    # Generate the next part of the story
    next_part = generate_text(story, max_new_tokens=50)  # Generate up to 50 new tokens

    # Update the story with the generated text
    story = next_part

    return story, story, ""  # Return the updated story, the state, and clear the user input

# Gradio interface
def create_gradio_interface():
    with gr.Blocks() as demo:
        gr.Markdown("# Interactive Storytelling with GPT-2")
        gr.Markdown("Guide the story by making choices at key moments. You have 5 steps to complete the story!")

        # State to store the story progression
        story_state = gr.State("")

        # Textbox to display the story
        story_display = gr.Textbox(
            label="Story",
            lines=10,
            interactive=False,
            value="Once upon a time, in a small village, there lived a brave adventurer named Alex. One day, Alex decided to explore the mysterious forest nearby."
        )

        # Textbox for user input
        user_input = gr.Textbox(label="Your Choice", placeholder="What do you want to do next?")

        # Button to submit the choice
        submit_button = gr.Button("Submit Choice")

        # Output and state update
        submit_button.click(
            fn=interactive_storytelling,
            inputs=[user_input, story_state],
            outputs=[story_display, story_state, user_input]
        )

        # Clear button to restart the story
        clear_button = gr.Button("Restart Story")
        clear_button.click(
            fn=lambda: ("Once upon a time, in a small village, there lived a brave adventurer named Alex. One day, Alex decided to explore the mysterious forest nearby.", "", ""),  # Reset story, state, and user input
            inputs=[],
            outputs=[story_display, story_state, user_input]
        )

    return demo

# Create and launch the Gradio interface
if __name__ == "__main__":
    interface = create_gradio_interface()
    interface.launch()
```

```python
import torch
from transformers import GPT2Tokenizer, GPT2LMHeadModel
import gradio as gr
import random

# Load pre-trained GPT-2 model and tokenizer
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')

# Function to generate text based on a prompt
def generate_text(prompt, max_new_tokens=50, temperature=0.7, no_repeat_ngram_size=2):
    input_ids = tokenizer.encode(prompt, return_tensors='pt')
    output = model.generate(
        input_ids,
        max_new_tokens=max_new_tokens,  # Use max_new_tokens instead of max_length
        temperature=temperature,
        no_repeat_ngram_size=no_repeat_ngram_size,
        pad_token_id=tokenizer.eos_token_id
    )
    return tokenizer.decode(output[0], skip_special_tokens=True)

# Function to generate a random story introduction
def generate_random_intro():
    intros = [
        "In a hidden kingdom beyond the mountains, an apprentice wizard named Elyon discovered a forbidden spellbook.",
        "Deep within an ancient jungle, an explorer named Zane stumbled upon a mysterious glowing artifact.",
        "On a stormy night, a lone traveler named Seraphina took shelter in a seemingly abandoned castle.",
        "In the bustling city of Eldoria, a young detective named Jace received an anonymous letter leading to a dangerous secret.",
        "Beneath the ocean’s depths, a mermaid named Nereida found an ancient key that unlocked a lost civilization."
    ]
    return random.choice(intros)

# Function to handle the interactive storytelling
def interactive_storytelling(user_choice, story_state=""):
    if not story_state:
        # Start a new story with a random introduction
        story = generate_random_intro()
    else:
        story = story_state

    # Update the story with the user's choice
    story += f" {user_choice}. "

    # Generate the next part of the story
    next_part = generate_text(story, max_new_tokens=50)

    # Update the story with the generated text
    story = next_part

    return story, story, ""  # Return the updated story, the state, and clear the user input

# Gradio interface
def create_gradio_interface():
    with gr.Blocks(theme=gr.themes.Base(), css="""
        body { background-color: #121212; color: #ffffff; }
        .gradio-container { background-color: #1e1e1e; }

        /* Ensure all textboxes are black with white text */
        textarea, input {
            background-color: #000000 !important;  /* Black background */
            color: #ffffff !important;  /* White text */
            border: 1px solid #555555;  /* Slight border */
        }

        button:hover { background-color: #444444; }
        h1, h2, h3, h4, h5, h6, p { color: #f8f8f8 !important; }
    """) as demo:
        gr.Markdown("""<h1 style='color: #ffcc00;'>Interactive Storytelling with GPT-2</h1>""")
        gr.Markdown("""<p style='color: #ffffff;'>Guide the story by making choices at key moments. Each restart begins a new adventure!</p>""")

        # State to store the story progression
        story_state = gr.State("")

        # Textbox to display the story
        story_display = gr.Textbox(
            label="Story",
            lines=10,
            interactive=False,
            value=generate_random_intro()
        )

        # Textbox for user input
        user_input = gr.Textbox(label="Your Choice", placeholder="What do you want to do next?")

        # Button to submit the choice
        submit_button = gr.Button("Submit Choice")

        # Output and state update
        submit_button.click(
            fn=interactive_storytelling,
            inputs=[user_input, story_state],
            outputs=[story_display, story_state, user_input]
        )

        # Clear button to restart the story with a new intro
        clear_button = gr.Button("Restart Story")
        clear_button.click(
            fn=lambda: (generate_random_intro(), "", ""),  # Reset story, state, and user input with a new story intro
            inputs=[],
            outputs=[story_display, story_state, user_input]
        )

    return demo

# Create and launch the Gradio interface
if __name__ == "__main__":
    interface = create_gradio_interface()
    interface.launch()
```
